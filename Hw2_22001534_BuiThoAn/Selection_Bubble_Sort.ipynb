```python
import time
import random
import matplotlib.pyplot as plt
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
```

```python
def measure_time(func, arr):
    start = time.time()
    func(arr)
    end = time.time()
    return end - start
```

```python
sizes = [100, 500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
selection_times = []
bubble_times = []

for size in sizes:
    arr = [random.randint(0, 10000) for _ in range(size)]
    selection_times.append(measure_time(selection_sort, arr))
    print(f"Selection sort {size} elements: {selection_times[-1]} s")
    bubble_times.append(measure_time(bubble_sort, arr))
    print(f"Bubble sort {size} elements: {bubble_times[-1]} s")

plt.plot(sizes, selection_times, marker='o', linestyle='-', label='Selection Sort')
plt.plot(sizes, bubble_times, marker='s', linestyle='-', label='Bubble Sort')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Selection Sort vs Bubble Sort')
plt.legend()
plt.grid()
plt.show()
```
